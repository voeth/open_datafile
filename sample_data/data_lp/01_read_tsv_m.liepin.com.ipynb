### 清数据：用函数

```python
# ## 990-990万 保密 面议
def clean_salary(dlist):
    dlist_output = []
    for data in dlist:
        if '万' in data:
            datac = data.replace('万','')  #data.strip('万')
            datac = datac.replace('以上','-')  #data.strip('万')
            datac = datac.replace('以下','')  #data.strip('万')
            datao = [int(x) if x !='' else 0 for x in datac.split('-')]
        elif '保密' in data or '面议' in data:
            datao = [None,None]
        dlist_output.append(datao)
        
    return(dlist_output)           
    
```

```python
clean_salary(list(df_all.salary)[0:40])
```

```python
df_s = pd.DataFrame(clean_salary(list(df_all.salary)))
df_s.columns=["下","上"]
df_s.head()
```

### 清数据：用pandas建新栏位

```python
# df_all["新"] = df_all.salary
df_all["新"] = ["" if "保密" in x or "面议" in x else x for x in list(df_all.salary)]
```

```python
df_all["新2"] =  [x.replace('万','').replace('以上','').replace('以下','-') for x in df_all.新]
```

```python
df_all["下"] = [int(x.split('-')[0]) if '-' in x else x for x in df_all["新2"]]
df_all["下"].head()
```

```python
df_all["上"] = [x.split('-')[1] if '-' in x else x for x in df_all["新2"]]
df_all["上"].head()
```

```python
df_all["上"] = [None if x=='' else int(x) for x in df_all["上"]]
df_all["上"].head()
```

```python
df_all.describe()
```
